The goal is to handle outliers (extremely high or low values) in a dataset without completely removing them. This is done using a method called IQR filtering combined with capping, which allows us to limit the influence of extreme values while keeping useful information.
Step-by-Step Explanation
Step 1: Calculate IQR for Each Feature
Calculate Q1 (25th percentile) and Q3 (75th percentile):

Q1 (25th percentile): This is the value below which 25% of the data falls.
Step 2: Cap Outliers
Replace values below the lower bound:

If a data point is below the calculated lower bound, we replace it with the value of the lower bound itself.
This process is known as capping the outliers at the lower end.
Replace values above the upper bound:

If a data point is above the calculated upper bound, we replace it with the value of the upper bound.
This is called capping the outliers at the upper end.Q3 (75th percentile): This is the value below which 75% of the data falls.
Calculate IQR (Interquartile Range):

The Interquartile Range (IQR) is the difference between Q3 and Q1.
Formula:
IQR=Q3−Q1
The IQR measures the spread of the middle 50% of the data, and it is useful in identifying outliers.
Determine Lower and Upper Bounds:

To identify outliers, we calculate the lower bound and upper bound using the following formulas:
Lower Bound: 

Q1−1.5×IQR
Upper Bound: 

Q3+1.5×IQR
Values below the lower bound or above the upper bound are considered potential outliers.

```python
def cap_outliers(df, column):
    # Calculate Q1 and Q3
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculate lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Cap outliers
    df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    
# Apply to the selected columns
columns_to_cap = ['price', 'reviews_per_month', 'number_of_reviews']

for column in columns_to_cap:
    cap_outliers(df, column)

# Check the updated DataFrame for outliers
print(df[columns_to_cap].describe())
```

```python
from sklearn.model_selection import train_test_split
features = df[['number_of_reviews', 'reviews_per_month', 'bedrooms', 'availability_365']]
target = df['price']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error: {mae:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f'R-squared: {r2:.2f}')


```

Mean Absolute Error (MAE: 65.03):
On average, the predictions are off by approximately $65.03. While this gives an idea of the prediction error, it can be misleading without context on the actual prices.
Root Mean Squared Error (RMSE: 83.22):
The RMSE is generally a more informative metric since it gives higher weight to larger errors. An RMSE of $83.22 suggests that there are some substantial prediction errors.
R-squared (0.20):
An R-squared value of 0.20 indicates that the model explains only 20% of the variance in the target variable (price). This is quite low, suggesting that other factors might be influencing the price that are not captured in your model.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, classification_report, accuracy_score

# Assuming df is your DataFrame
# Step 1: Define Price Categories
bins = [0, 100, 200, 300, 400, np.inf]  # Define your price bins
labels = ['Very Low', 'Low', 'Medium', 'High', 'Very High']  # 5 labels for 5 bins

df['price_category'] = pd.cut(df['price'], bins=bins, labels=labels, right=False)  # right=False to include left edge

# Step 2: Prepare Features and Targets for Regression and Classification
features = df[['number_of_reviews', 'reviews_per_month', 'bedrooms', 'availability_365']]
target_regression = df['price']  # For regression
target_classification = df['price_category']  # For classification

# Step 3: Split the Data for Regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(features, target_regression, test_size=0.2, random_state=42)

# Step 4: Train and Evaluate the Regression Model
regression_model = LinearRegression()
regression_model.fit(X_train_reg, y_train_reg)

# Predictions
y_pred_reg = regression_model.predict(X_test_reg)

# Regression Metrics
mae = mean_absolute_error(y_test_reg, y_pred_reg)
rmse = mean_squared_error(y_test_reg, y_pred_reg, squared=False)  # RMSE
r2 = r2_score(y_test_reg, y_pred_reg)

print(f'Regression Metrics:')
print(f'Mean Absolute Error: {mae:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f'R-squared: {r2:.2f}')

# Step 5: Split the Data for Classification
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(features, target_classification, test_size=0.2, random_state=42)

# Step 6: Train and Evaluate the Classification Model
from sklearn.linear_model import LogisticRegression

classification_model = LogisticRegression(max_iter=1000)
classification_model.fit(X_train_cls, y_train_cls)

# Predictions
y_pred_cls = classification_model.predict(X_test_cls)

# Classification Metrics
accuracy = accuracy_score(y_test_cls, y_pred_cls)
report = classification_report(y_test_cls, y_pred_cls)

print(f'Classification Metrics:')
print(f'Accuracy: {accuracy:.2f}')
print('Classification Report:')
print(report)
```
